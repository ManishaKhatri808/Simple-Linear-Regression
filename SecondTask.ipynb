```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
```

```python
iris_data=pd.read_csv('D:\Simple-Unsupervised-Prediction-System\Second-Task\Iris.csv')
```

```python
x = iris_data.iloc[:, :-1]
y = iris_data.iloc[:, -1].values  
```

```python
#finding optimum number of clusters
wcss = []

#preprocessing the data
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y)

for i in range(1,11):
    kmeans = KMeans(n_clusters=i, init= 'k-means++', max_iter=300, n_init = 10, random_state=0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)

# Plotting the results onto a line graph to observe 'the elbow'

plt.plot(range(1,11), wcss)
plt.title('The elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show() 
```

It is concluded that there are 3 clusters from the above visualization

```python
x_new = iris_data.iloc[:, :-1].values

# Applying kmeans to the dataset
kmeans = KMeans(n_clusters = 3, init = 'k-means++',
                max_iter = 300, n_init = 10, random_state = 0)
y_kmeans = kmeans.fit_predict(x_new)

# Visualising the clusters on the first two columns
plt.scatter(x_new[y_kmeans == 0, 0], x_new[y_kmeans == 0, 1], s = 100, c = 'red', label = 'Iris-setosa')
plt.scatter(x_new[y_kmeans == 1, 0], x_new[y_kmeans == 1, 1], s = 100, c = 'blue', label = 'Iris-versicolour')
plt.scatter(x_new[y_kmeans == 2, 0], x_new[y_kmeans == 2, 1], s = 100, c = 'green', label = 'Iris-virginica')

# Plotting the centroids of the clusters
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1],s = 100, c = 'yellow', label = 'Centroids')

plt.legend()
```

By using the pair plot, we can observe the distributions and relationships of the different features of each species, as well as identify any possible clusters or patterns in the data. The plot can help in determining the most important features that can be used to differentiate between the different species of Iris flowers.
